## Train the model

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_absolute_error

max_depth = 100
```

```python
rfc = RandomForestRegressor(n_estimators=200, max_depth=max_depth,
                                random_state=2)
rfc.fit(X_train, y_train)
print( 'The score of Random Forest Classifier: ' + str(mean_absolute_error(y_val, rfc.predict(X_val))) )
```

## Show the importance of features

```python
print('Sorted importance of features: \n')
print(sorted(zip(map(lambda x: round(x, 4), rfc.feature_importances_), X.columns), reverse=True))
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
print( 'The score of Default Random Forest Classifier: ' + str(mean_absolute_error(y_val, model.predict(X_val))) )
```

```python
rfc_multi = MultiOutputRegressor(RandomForestRegressor(n_estimators=200,
                                                          max_depth=max_depth,
                                                          random_state=0))
rfc_multi.fit(X_train, y_train)
print( 'The score of Multiple Random Forest Classifier: ' + str(mean_absolute_error(y_val, rfc_multi.predict(X_val))) )
```

```python
from sklearn.svm import SVC
linear_svm = MultiOutputRegressor(SVC(kernel = 'linear'))
linear_svm.fit(X_train, y_train)
print( 'The score of linear SVM: ' + str(mean_absolute_error(y_val, linear_svm.predict(X_val))) )
```

```python
rbf_svm = MultiOutputRegressor(SVC(kernel = 'rbf'))
rbf_svm.fit(X_train, y_train)
print( 'The score of rbf SVM: ' + str(mean_absolute_error(y_val, rbf_svm.predict(X_val))) )
```

```python
poly_svm = MultiOutputRegressor(SVC(kernel = 'poly'))
poly_svm.fit(X_train, y_train)
print( 'The score of poly SVM: ' + str(mean_absolute_error(y_val, poly_svm.predict(X_val))) )
```

```python
sigmoid_svm = MultiOutputRegressor(SVC(kernel = 'sigmoid'))
sigmoid_svm.fit(X_train, y_train)
print( 'The score of sigmoid SVM: ' + str(mean_absolute_error(y_val, sigmoid_svm.predict(X_val))) )
```

```python
import xgboost as xgb
XGBoost = MultiOutputRegressor(xgb.XGBRegressor(n_estimators=1000, max_depth=100, learning_rate=0.01, \
                           subsample=1.0, gamma=0.5, reg_alpha=0.1, reg_lambda=1.0))
XGBoost.fit(X_train, y_train)
print( 'The score of XGB: ' + str(mean_absolute_error(y_val, XGBoost.predict(X_val))) )
```

```python
params = {'num_leaves': 21,
         'min_data_in_leaf': 20,
         'objective':'regression',
         'learning_rate': 0.001,
         'max_depth': 108,
         "boosting": "gbdt",
         "feature_fraction": 0.91,
         "bagging_freq": 1,
         "bagging_fraction": 0.91,
         "bagging_seed": 42,
         "metric": 'mae',
         "lambda_l1": 0.1,
         "verbosity": -1,
         "random_state": 42}
```

```python
import lightgbm as lgb
lightGBM = MultiOutputRegressor(lgb.LGBMRegressor(**params, n_estimators=80000, n_jobs=-1))
lightGBM.fit(X_train, y_train)
print( 'The score of lightGBM: ' + str(mean_absolute_error(y_val, lightGBM.predict(X_val))) )
```
